```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_data = "mouse_drug_data.csv"
trial_data = "clinical_trial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_data)
trial_df = pd.read_csv(trial_data)

# Combine the data into a single dataset
combo_df = trial_df.merge(mouse_df, on = "Mouse ID")


# Display the data table for preview
combo_df.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

#make lists of unique timepoints and drug names
timepoints = combo_df["Timepoint"].unique()
timepoints = timepoints.tolist()

drug_names = combo_df["Drug"].unique()
drug_names = drug_names.tolist()

#create blank lists to hold data
time_list = []
drug_list = []
volume_list = []

#for loop to grab data that will:
# 1. loop through each unique timepoint 0-45
# 2. loop through each unique drug name for each timepoint
# 3. search through the combined dataframe to grab the mean for each timepoint/drug combination
# 4. append the information to new lists that can be combined into a new dataframe
for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Tumor Volume (mm3)"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].mean()
        volume_list.append(z)
          
# Convert to DataFrame
df_1 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Tumor Volume (mm3)":volume_list}

df_1 = pd.DataFrame(df_1)



# Preview DataFrame (sort by drug and timepoint)
df_1 = df_1.sort_values(["Drug","Timepoint"])

#lazy workaround because renaming messed up some code further down
df_6 = df_1

df_1.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# reset lists
time_list = []
drug_list = []
volume_list = []

# for loop same as above except this time get the standard error
for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Tumor Volume (mm3)"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].sem()
        volume_list.append(z)
          

# Convert to DataFrame
df_2 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Tumor Volume (mm3)":volume_list}

df_2 = pd.DataFrame(df_2)


# Preview DataFrame & sort

df_2 = df_2.sort_values(["Drug","Timepoint"])
df_2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# pivot the dataframes using timepoint as the index and drugs as the columns
df_1 = df_1.pivot_table(index ="Timepoint", columns ="Drug", values ="Tumor Volume (mm3)")
df_2 = df_2.pivot_table(index ="Timepoint", columns ="Drug", values ="Tumor Volume (mm3)")

```

```python
# Preview that Reformatting worked
df_1.head()
```

```python
# Generate the Plot (with Error Bars)

# show mean dataframe as plot and use standard error dataframe for error bars
plot_1 = df_1.plot(figsize = (20, 15), fontsize = 15, yerr = df_2, marker = "o")

plot_1.grid()

#format labels
plot_1.set_xlabel("Time (Days)", fontsize = 25)
plot_1.set_ylabel("Tumor Size", fontsize = 25)
plot_1.set_title("Tumor Volume Over Time", fontsize = 30)

# Save the Figure
plt.savefig('tumor_volume_over_time.png')
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint

#reset lists again
time_list = []
drug_list = []
meta_list = []

#same for loop, this time get the mean metastatic site data
for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Metastatic Sites"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].mean()
        meta_list.append(z)
          
# Convert to DataFrame
df_3 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Metastatic Sites":meta_list}

df_3 = pd.DataFrame(df_3)


# Preview DataFrame
df_3 = df_3.sort_values(["Drug","Timepoint"])
df_3.head()

```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

#new lists
time_list = []
drug_list = []
mouse_list = []

#for loop again, get count of mice existing at each timepoint for each drug
#assume that if they're no longer in the study, they are dead
for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Mouse ID"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].count()  
        mouse_list.append(z)
        

# Convert to DataFrame

df_5 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Surviving Mice":mouse_list}

df_5 = pd.DataFrame(df_5)


# Preview DataFrame

df_5 = df_5.sort_values(["Drug","Timepoint"])
df_5.head()

```

```python
# Store all Relevant Percent Changes into a Tuple
pc_df = {"Drug":names,
        "Percent Change":pc}

pc_df = pd.DataFrame(pc_df)

plot_4 = pc_df.plot(kind="bar",figsize = (15, 10), fontsize = 15, color=[np.where(pc_df["Percent Change"]>0, 'r', 'g')])

# Orient widths. Add labels, tick marks, etc. 
plot_4.set_xticklabels(names, rotation=45)

plot_4.set_title("Percent Change in Tumor Volume Over Time", fontsize = 30)
plot_4.set_xlabel("Drug Name", fontsize = 20)
plot_4.set_ylabel("Percent Change over 45 Days", fontsize = 20)


#label the percentages of changes
rounded = [round(x) for x in pc]
percents = [f"{y}%" for y in rounded]
x_values = [-.22, .83, 1.83, 2.83, 3.83, 4.83, 5.83, 6.83, 7.83, 8.83]

plot_4.text(x_values[0], rounded[0]-3, percents[0], fontsize=15)
plot_4.text(x_values[1], rounded[1]+1, percents[1], fontsize=15)
plot_4.text(x_values[2], rounded[2]+1, percents[2], fontsize=15)
plot_4.text(x_values[3], rounded[3]+1, percents[3], fontsize=15)
plot_4.text(x_values[4], rounded[4]+1, percents[4], fontsize=15)
plot_4.text(x_values[5], rounded[5]+1, percents[5], fontsize=15)
plot_4.text(x_values[6], rounded[6]+1, percents[6], fontsize=15)
plot_4.text(x_values[7], rounded[7]-3, percents[7], fontsize=15)
plot_4.text(x_values[8], rounded[8]+1, percents[8], fontsize=15)
plot_4.text(x_values[9], rounded[9]+1, percents[9], fontsize=15)

#get rid of useless legend
plot_4.get_legend().remove()

# Save the Figure
plt.savefig('tumor_pc_over_time.png')

# Show the Figure
plt.show()
```
