```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_data = "mouse_drug_data.csv"
trial_data = "clinical_trial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_df = pd.read_csv(mouse_data)
trial_df = pd.read_csv(trial_data)

# Combine the data into a single dataset
combo_df = trial_df.merge(mouse_df, on = "Mouse ID")


# Display the data table for preview
combo_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
timepoints = combo_df["Timepoint"].unique()
timepoints = timepoints.tolist()

drug_names = combo_df["Drug"].unique()
drug_names = drug_names.tolist()

time_list = []
drug_list = []
volume_list = []

for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Tumor Volume (mm3)"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].mean()
        volume_list.append(z)
          
# Convert to DataFrame
df_1 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Tumor Volume (mm3)":volume_list}

df_1 = pd.DataFrame(df_1)


# Preview DataFrame
df_1 = df_1.sort_values(["Drug","Timepoint"])
df_1.head(20)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

time_list = []
drug_list = []
volume_list = []

for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Tumor Volume (mm3)"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].sem()
        volume_list.append(z)
          

# Convert to DataFrame

df_2 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Tumor Volume (mm3)":volume_list}

df_2 = pd.DataFrame(df_2)


# Preview DataFrame

df_2 = df_2.sort_values(["Drug","Timepoint"])
df_2.head(20)
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_3 = df_1.pivot_table(index ="Timepoint", columns ="Drug", values ="Tumor Volume (mm3)")
df_4 = df_2.pivot_table(index ="Timepoint", columns ="Drug", values ="Tumor Volume (mm3)")

# Preview that Reformatting worked
df_4.head()
```

```python
# Generate the Plot (with Error Bars)


plot_1 = df_3.plot(figsize = (20, 15), fontsize = 15, yerr = df_4, marker = "o")


plot_1.set_xlabel("Time (Days)", fontsize = 25)
plot_1.set_ylabel("Tumor Size", fontsize = 25)
plot_1.set_title("Tumor Volume Over Time", fontsize = 30)

# Save the Figure
plt.savefig('tumor_volume_over_time.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint

time_list = []
drug_list = []
meta_list = []

for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Metastatic Sites"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].mean()
        meta_list.append(z)
          
# Convert to DataFrame
df_5 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Metastatic Sites":meta_list}

df_5 = pd.DataFrame(df_5)


# Preview DataFrame
df_5 = df_5.sort_values(["Drug","Timepoint"])
df_5.head(20)

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
time_list = []
drug_list = []
meta_list = []

for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Metastatic Sites"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].sem()
        meta_list.append(z)
          
# Convert to DataFrame
df_6 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Metastatic Sites":meta_list}

df_6 = pd.DataFrame(df_6)


# Preview DataFrame
df_6 = df_6.sort_values(["Drug","Timepoint"])
df_6.head(20)

```

```python
# Minor Data Munging to Re-Format the Data Frames
df_7 = df_5.pivot_table(index ="Timepoint", columns ="Drug", values ="Metastatic Sites")
df_8 = df_6.pivot_table(index ="Timepoint", columns ="Drug", values ="Metastatic Sites")


# Preview that Reformatting worked
df_7
```

```python
# Generate the Plot (with Error Bars)
plot_2 = df_7.plot(figsize = (20, 15), fontsize = 15, yerr = df_8, marker = "o")


plot_2.set_xlabel("Time (Days)", fontsize = 25)
plot_2.set_ylabel("Metastatic Sites", fontsize = 25)
plot_2.set_title("Quantity of Metastatic Sites Over Time", fontsize = 30)

# Save the Figure
plt.savefig('meta_sites_over_time.png')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
time_list = []
drug_list = []
mouse_list = []

for x in timepoints:
    for y in drug_names:
        drug_list.append(y)
        time_list.append(x)
        z = combo_df["Mouse ID"].loc[(combo_df["Drug"] == y) & (combo_df["Timepoint"] == x)].count()  
        mouse_list.append(z)
        

# Convert to DataFrame

df_9 = {"Drug":drug_list,
       "Timepoint":time_list,
       "Surviving Mice":mouse_list}

df_9 = pd.DataFrame(df_9)


# Preview DataFrame

df_9 = df_9.sort_values(["Drug","Timepoint"])
df_9.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
df_10 = df_9.pivot_table(index ="Timepoint", columns ="Drug", values ="Surviving Mice")

df_10
# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)

plot_3 = df_10.plot(figsize = (20, 15), fontsize = 15, marker = "o")


plot_3.set_xlabel("Time (Days)", fontsize = 25)
plot_3.set_ylabel("Surviving Mice", fontsize = 25)
plot_3.set_title("Quantity of Surviving Mice Over Time", fontsize = 30)

# Save the Figure
plt.savefig('survivors_over_time.png')

# Show the Figure
plt.show()
```
